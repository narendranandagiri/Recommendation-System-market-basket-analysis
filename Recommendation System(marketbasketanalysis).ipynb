```python
import pandas as pd
df = pd.read_csv('/kaggle/input/onlineretail/onlineRetail.csv')
df.head()
```

```python
df['Invoice'].unique()
```

```python
df[df['Invoice'].str.contains('C')]
```

* Invoive number with prefic c represents return items
* So we can remove such items from our analysis for recommendation system

```python
df1 = df[~df['Invoice'].str.contains('C')]
df1.head()
```

****
* Using Invoice number we can identify items in one basket
****

```python
df1.groupby(['Invoice','Description'])['Quantity'].sum()
```

* Remove items with prefic 'A' as it is to adjust bad debts

```python
df1 = df1[~df1['Invoice'].str.contains('A')]
df1.head()
```

```python
df1['Description'] = df1['Description'].str.strip()
df1['Description'] = df1['Description'].str.lower()
```

```python
df1.groupby(['Invoice','Description'])['Quantity'].sum()
```

```python
dfx = df1.groupby(['Invoice','Description'])['Quantity'].sum().unstack().fillna(0)
dfx.head()
```

```python
dfx['pink cherry lights']
```

* Convert all non-zeros into 1
* Identify most frequent itemsets
* Apply Aprori algorithm to generate recommenadation system

```python
basket = dfx.apply(lambda x : x > 0)
basket
```

* most frequent items

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(basket , use_colnames = True, min_support = 0.01)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules1 = association_rules(frequent_itemsets, metric='lift' , min_threshold=1)
rules1
```

```python
rules1.columns
```

```python
r1 = rules1[['antecedents', 'consequents' , 'support', 'confidence', 'lift']]
r1.head()
```

```python
r1['antecedents'] = r1['antecedents'].apply(lambda x : list(x))
r1['consequents'] = r1['consequents'].apply(lambda x : list(x))
```

```python
r1.to_csv('rules1_invoice.csv')
```

* Create a recomendation mechanism for USA

```python
df_usa = df1[df1['Country'] == 'USA']
df_usa.head()
```

* Based on invoice group the items

```python
df_usa_basket = df_usa.groupby(['Invoice','Description'])['Quantity'].sum().unstack().fillna(0)
df_usa_basket
```

```python
df_usa_basket = df_usa_basket.applymap(lambda x : x > 0)
df_usa_basket.head()
```

* find out most frquent itemsets

```python
from mlxtend.frequent_patterns import apriori
most_frequent_items_usa = apriori(df_usa_basket , use_colnames = True , 
                                  min_support=0.08)
most_frequent_items_usa
```

* Use apriori to identify best possible antecedents and consequents

```python
from mlxtend.frequent_patterns import association_rules
high_association_usa = association_rules(most_frequent_items_usa , metric = 'lift' , 
                                         min_threshold = 1)
high_association_usa.head()
```

```python
high_association_usa.columns
```

```python
r2_usa = high_association_usa[['antecedents', 'consequents' , 'support', 
                               'confidence', 'lift']]
r2_usa
```

```python
r2_usa['antecedents'] = r2_usa['antecedents'].apply(lambda x : list(x))
r2_usa['consequents'] = r2_usa['consequents'].apply(lambda x : list(x))
```

```python
r2_usa.head()
```

```python
r2_usa.to_csv("r2_usa.csv")
```

* Create recomendation mechanism using Customer ID

```python
df1_cusId = df1.groupby(['Customer ID','Description'])['Quantity'].sum().unstack().fillna(0)
df1_cusId.head()
```

```python
cust_basket = df1_cusId.applymap(lambda x : x > 0)
cust_basket.head()
```

```python
from mlxtend.frequent_patterns import apriori
most_frequent_items_ = apriori(cust_basket , use_colnames = True , min_support=0.07 )
most_frequent_items_
```

```python
from mlxtend.frequent_patterns import association_rules
high_association_cust_basket = association_rules(most_frequent_items_ , 
                                                 metric = 'lift' , min_threshold = 1)
high_association_cust_basket.head()
```

```python
r2_custId = high_association_cust_basket[['antecedents', 'consequents' , 
                                    'support', 'confidence', 'lift']]
r2_custId
```

```python
r2_custId['antecedents'] = r2_custId['antecedents'].apply(lambda x : list(x))
r2_custId['consequents'] = r2_custId['consequents'].apply(lambda x : list(x))
```

```python
r2_custId.head()
```

```python
r2_custId.to_csv('r2_custId.csv')
```
